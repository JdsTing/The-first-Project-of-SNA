```python
#匯入套件
import pandas as pd
import jieba
import jieba.analyse

#基礎設定
jieba.set_dictionary("005 dict.txt")
jieba.load_userdict("005 dict_user.txt")  #2019新聞議題排行 文字   https://dailyview.tw/
```

```python
#存成utf8_sig格式的csv檔
#artical_0901_1115school.to_csv("artical_0901_1115school.csv", encoding = "utf_8_sig")
#artical_0901_1115topic.to_csv("artical_0901_1115topic.csv", encoding = "utf_8_sig")
#artical_pop.to_csv("004_artical_pop.csv", encoding = "utf_8_sig")
# 讀入 csv 文字檔
csv_file = "004_artical_pop.csv"
input_artical = pd.read_csv(csv_file) 
input_artical.head()
```

```python
#斷詞
df = input_artical
xdf_cut=pd.DataFrame()
for i in range(len(df)):
    for x, w in jieba.analyse.extract_tags(df.loc[i]['content'], withWeight=True):
        xdf_cut=xdf_cut.append(pd.DataFrame([{'id': df.loc[i]['id'], 
                                              'Text': x, 
                                              'TF_IDF': w}]),
                               ignore_index=True)
xdf_cut.to_csv("005_xdf_cut_pop.csv", encoding = "utf_8_sig")
```
